```python
import sklearn.datasets as datasets
import pandas as pd
```

```python
# X, y = datasets.make_friedman2(n_samples=40000, noise=0, random_state=42)
X, y = datasets.make_regression(n_samples=40000, noise=0, n_features=1, random_state=42)
df = pd.DataFrame(X)
df['target'] = y

# insert first column named 'id', which is just the index
df.insert(0, 'id', df.index)
```

```python
df.head()
```

```python
# plot the distribution of the target
df['target'].hist()
```

```python
# plot target vs feature
df.plot.scatter(x=0, y='target')
```

```python
# save the dataset
df.to_csv('dummy_regression.csv', index=False)
```
